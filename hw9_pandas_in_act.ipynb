```python
import pandas as pd
cer = pd.read_csv("cereals.csv")
cer.head()
```

```python
# Сгруппируйте завтраки по значениям в столбце Manufacturer (Произво- дитель)
manf = cer.groupby("Manufacturer")
manf
```

```python
# 2 Определите общее количество групп и завтраков в каждой группе
len(manf)
```

```python
# Извлеките список завтраков, выпускаемых производителем Nabisco
manf.get_group('Nabisco')
```

```python
# Вычислите средние значения по столбцам Calories, Fiber и Sugars для каж- дого производителя
manf["Calories","Fiber","Sugars"].mean()
```

```python
# Найдите максимальное значение в столбце Sugars для каждого произво- дителя
manf['Sugars'].max()
```

```python
# Найдите минимальное значение в столбце Fiber для каждого производителя
manf['Fiber'].min()
```

```python
# Найдите завтраки с наименьшим количеством сахара (столбец Sugars)
# по производителям и верните их в виде нового DataFrame

def small_sugar(ddf):
  return ddf.nsmallest(1, 'Sugars')

manf.apply(small_sugar)
```
