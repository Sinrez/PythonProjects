* Напишите функцию remove_items(L, i), которая уберёт все вхождения объекта i из листа L

(Метод .remove(i) убирает только один инстанс, и если его нет выбрасывает ошибку, так что функция не такая простая)

```python
# Ваш код
def remove_items(L, i):
  try:
    if not isinstance(L, list):
      raise ValueError('Передан не список!')
    return [l for l in L if l != i]
  except ValueError as va:
    print(f'{va}')
  except Exception as ex:
    print(f'Ошибка {ex}')

L = [1, 2, 3, 4, 1, 2, 3, 1, 2]
print("Принято" if remove_items(L, 1) == [2, 3, 4, 2, 3, 2] else "Что-то пошло не так")
# print("Принято" if remove_items('strrringgg', 1) == [2, 3, 4, 2, 3, 2] else "Что-то пошло не так")
```

* Реализуйте функцию reverse_dict(D), которая возвращает "развёрнутый" словарь (превращает его ключи в значения и значения в ключи). В случае если у двух ключей одинаковое значение - вместо этого верните None

В этом пункте верим что все значения хэшируются, если нет - пусть вылетает ошибка.

```python
# Ваш код
def reverse_dict(D):
    try:
      if not isinstance(D, dict):
        raise ValueError('Передан не словарь!')
      if len(set(D.values())) == 1:
          return None
      else:
          return {value: key for key, value in D.items() if key != value}
    except ValueError as va:
      return(f'{va}')
    except Exception as ex:
      return(f'Ошибка {ex}')
print(reverse_dict({1:"x", 2:"y", 3:"z"}))
print(reverse_dict({1:"x", 2:"x", 3:"x"}))
# print(reverse_dict('ssssttriiiiingggg'))
```

* Напишите генератор kollats(x) [последовательности Коллатца](https://ru.wikipedia.org/wiki/%D0%93%D0%B8%D0%BF%D0%BE%D1%82%D0%B5%D0%B7%D0%B0_%D0%9A%D0%BE%D0%BB%D0%BB%D0%B0%D1%82%D1%86%D0%B0#:~:text=%D0%91%D0%B5%D1%80%D1%91%D0%BC%20%D0%BB%D1%8E%D0%B1%D0%BE%D0%B5%20%D0%BD%D0%B0%D1%82%D1%83%D1%80%D0%B0%D0%BB%D1%8C%D0%BD%D0%BE%D0%B5%20%D1%87%D0%B8%D1%81%D0%BB%D0%BE%20n,%D0%B8%D0%BB%D0%B8%20%D0%BF%D0%BE%D0%B7%D0%B4%D0%BD%D0%BE%20%D0%BC%D1%8B%20%D0%BF%D0%BE%D0%BB%D1%83%D1%87%D0%B8%D0%BC%20%D0%B5%D0%B4%D0%B8%D0%BD%D0%B8%D1%86%D1%83.), который возвращает следующий элимент начиная с x

*Коротко: Последовательность начинается с x, и каждый следующий элемент равен 3x+1 если прошлый элемент чётный, в противном случае следующий элемент это x/2*

*Например генератор kollats(3) должен выдавать:*
3 10 5 16 8 4 2 1


```python
# Ваш код
def collatz(x):
    yield x
    while x != 1:
        if x % 2 == 0:
            x = x // 2
        else:
            x = 3 * x + 1
        yield x

for element in collatz(3):
    print(element, end=',')
```

* Напишите генератор only_even(X), который берёт на вход контейнер или генератор и возвращает из него только чётные по порядку элементы.

```python
# Ваш код
def only_even(X):
  for x in X:
    if x % 2 == 0:
      yield x

for x in only_even([0, 1, 2, 3, 4, 5, 6, 7]):
    print(x, end=", ")
print()

def cubes():
    i = 0
    while True:
        yield i ** 3
        i += 1

for i, x in enumerate(only_even(cubes())):
    if x > 1000000:
        break
    print(x, end=", ")
```

* Напишите генератор cloner(x), который берёт на вход контейнер, и возвращает первый элимент один раз, второй - два раза, третий - три раза и так далее.

```python
# Ваш код
def cloner(x):
  try:
    for i in range(len(x)):
      yield str(x[i]) * (i+1)
  except IndexError:
    exit()
  except Exception as ex:
    print(f'Ошибка {ex}')

for x in cloner([0, 1, 2, 3, 4, 5, 6, 7]):
    print(x, end=", ")
print()
```

* Выведите на экран треугольник Паскаля до n-ного ряда.
Треугольник Паскаяля это треугольник из чисел, у которого две стороны содержат единицы, а внутренние числа это суммы двух соседних с верхнего рядя

1

1 1

1 2 1

1 3 3 1

1 4 6 4 1

Здесь ряд 1 3 3 1 получился как 1, 1+2, 2+1, 1 из предыдущего

```python
# Ваш код. Можете либо просто вывести, либо сделать функцию pascal(N) и вызвать её
def print_pascal_triangle(n):
    pascal_triangle = [[1] * (i + 1) for i in range(n)]
    for i in range(2, n):
        for j in range(1, i):
            pascal_triangle[i][j] = pascal_triangle[i-1][j-1] + pascal_triangle[i-1][j]
    for row in pascal_triangle:
        print(' '.join(map(str, row)))

n = 5
print_pascal_triangle(n)
```

* Сгенерируйте лист случайных чисел от 0 до 1000000 (используйте библиотеку [Random](https://metanit.com/python/tutorial/6.1.php)).

Отсортируйте полученный массив таким образом, чтобы сначала в убывающем порядке шли все чётные числа, а потом в возростающем - нечётные.

```python
# Ваш код.
from random import randint

def rand_diff_sort_list():
  list_start = [randint(0, 1000000) for _ in range(0, 100000)]
  list_even = sorted([l for l in list_start if l % 2 == 0], reverse=True)
  list_not_even = sorted([l for l in list_start if l % 2 != 0])
  return list_even + list_not_even

if __name__ == '__main__':
  print(rand_diff_sort_list())
```

Написать генератор limit(generator, max_count). Возвращает не более max_count значений генератора generator.

```python
# Ваш код
def limit(generator, max_count):
    # реализация
    count = 0
    for val in generator:
        yield val
        count += 1
        if count >= max_count:
            break

for value in limit(cubes(), 10):
    print(value, end=' ')
```

Написать генератор all_elements(list). Возвращает все элементы списка list любой вложенности.

Указание: для проверки того, что объект итерируемый, можно проверить, что он наследник Iterable

```python
from collections.abc import Iterable

isinstance(1, Iterable)
```

```python
def all_elements(x):
    # реализация
    for i in x:
        if not isinstance(i, Iterable):
            yield i
        else:
            yield from all_elements(i)
              #  yield *i

values = [1, [2, 3], [4, [5, 6], [[[7]]]], 8]
for value in all_elements(values):
    print(value, end=' ')
```
