```python
import pandas as pd
investments = pd.read_csv("investments.csv")
investments.head()
```

```python
investments.nunique()
```

```python
# Создадим трехуровневый мультииндекс со столбцами Status, Funding Rounds и State
investments = investments.set_index(
             keys = ["Status", "Funding Rounds", "State"]
         ).sort_index()
investments.head()
```

```python
# Извлеките все строки с состоянием "Closed"
investments.loc['Closed'].head()
```

```python
# Извлеките все строки с состоянием "Acquired" и десятью раундами инве- стиций
investments.loc[('Acquired',10)].head()
```

```python
# Извлеките все строки с состоянием "Operating", шестью раундами инвести- ций и штатом "NJ"
investments.loc[('Operating',6,'NJ')]
```

```python
# Извлеките все строки с состоянием "Closed" и восемью раундами инвести- ций ИзвлекитетолькостолбецName
investments.loc[('Closed',8)]['Name']
```

```python
investments.loc[('Closed',8),('Name',)]
```

```python
# Извлеките все строки со штатом "NJ" вне зависимости от значений уровней
# Status и Funding Rounds
investments.xs(key = "NJ", level = "State").head()
```
