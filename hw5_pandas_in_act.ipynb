```python
import pandas as pd
```

```python
films = pd.read_csv(
    "/content/netflix.csv", parse_dates = ["date_added"]
)
films.head()
```

```python
# Добейтесь оптимального использования этим набором данных памяти и мак- симального удобства его применения

films["type"] = films["type"].astype("category")
films.info()
```

```python
# Найдите все строки с названием Limitless
films[films['title'] == 'Limitless']
```

```python
# Найдите все строки с режиссером Robert Rodriguez и типом Movie
pattern1 = films['director'] == 'Robert Rodriguez'
pattern2 = films['type'] == 'Movie'
films[pattern1 & pattern2]
```

```python
# Найдите все строки с датой добавления 2019-07-31 или режиссером Robert Altman
pattern1 = films['date_added'] == '2019-07-31'
pattern2 = films['director'] == 'Robert Altman'
films[pattern1 | pattern2]
```

```python
# Найдите все строки с режиссером Orson Welles, Aditya Kripalani или Sam Raimi
pattern1 = films['director'] == 'Sam Raimi'
pattern2 = films['director'] == 'Orson Welles'
pattern3 = films['director'] == 'Aditya Kripalani'
films[pattern1 | pattern2 | pattern3]
```

```python
# var2
directors = ['Sam Raimi','Orson Welles', 'Aditya Kripalani']
find_directors = films['director'].isin(directors)
films[find_directors]
```

```python
# Найдите все строки со значением столбца date_added между 1 мая и 1 июня # 2019 года
filter = films['date_added'].between(
    left = "2019-05-01",
    right = "2019-06-01"
)
films[filter]
```

```python
# Удалите все строки с NaN в столбце director
films['director'] = films['director'].dropna()
films.info()
```

```python
# Найдите все дни, в которые Netflix добавил в каталог только одну картину
films.drop_duplicates(subset= ["date_added"], keep = False)
```
