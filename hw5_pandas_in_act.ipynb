```python
import pandas as pd
```

```python
films = pd.read_csv(
    "/content/netflix.csv", parse_dates = ["date_added"]
)
films.head()
```

```python
# Добейтесь оптимального использования этим набором данных памяти и мак- симального удобства его применения

films["type"] = films["type"].astype("category")
films.info()
```

```python
# Найдите все строки с названием Limitless
films[films['title'] == 'Limitless']
```

```python
# Найдите все строки с режиссером Robert Rodriguez и типом Movie
pattern1 = films['director'] == 'Robert Rodriguez'
pattern2 = films['type'] == 'Movie'
films[pattern1 & pattern2]
```

```python
# Найдите все строки с датой добавления 2019-07-31 или режиссером Robert Altman
pattern1 = films['date_added'] == '2019-07-31'
pattern2 = films['director'] == 'Robert Altman'
films[pattern1 | pattern2]
```
