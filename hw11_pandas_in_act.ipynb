```python
import pandas as pd
# Преобразуйте строковые значения в столбцах start_time и stop_time в значения типа Timestamp
citi_bike = pd.read_csv("citibike.csv", parse_dates=["start_time", "stop_time"])
citi_bike.head()
```

```python
citi_bike.info()
```

Подсчитайте количество поездок, совершаемых по дням недели (поне- дельник, вторник и т д ) В какой будний день совершается больше всего велопоездок? Используйте столбец start_time в качестве отправной точки

```python

citi_bike["start_time"].dt.day_name().value_counts()
```

Подсчитайте количество поездок за каждую неделю в течение месяца Для это- го округлите все даты в столбце start_time до предыдущего или текущего понедельника Предположим, что каждая неделя начинается в понедельник и заканчивается в воскресенье Соответственно, первая неделя июня будет начинаться в понедельник 1 июня и заканчиваться в воскресенье 7 июня

```python
 days_away_from_monday = citi_bike["start_time"].dt.dayofweek
 citi_bike["start_time"] - pd.to_timedelta(
              days_away_from_monday, unit = "day"
)
```

```python
 dates_rounded_to_monday = citi_bike[
              "start_time"
          ] - pd.to_timedelta(days_away_from_monday, unit = "day")
```

```python
dates_rounded_to_monday.dt.date.value_counts()
```

Рассчитайте продолжительность каждой поездки и сохраните результаты в новый столбец duration

```python
citi_bike['duration'] = citi_bike['stop_time'] - citi_bike['start_time']
citi_bike.head()
```

```python
# Найдите среднюю продолжительность поездки
citi_bike['duration'].mean()
```

```python
# Извлеките из набора данных пять самых долгих поездок
citi_bike.sort_values('duration', ascending=False).head(5)
```

```python
# 2 способ
citi_bike.nlargest(n = 5, columns = "duration")
```
