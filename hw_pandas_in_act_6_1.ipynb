Наш набор данных покупателей включает столбец Address Адреса состоят из улицы, города, штата и почтового индекса Ваша задача: разбить эти четыре значения по четырем новым столбцам Street, City, State и Zip, после чего уда- лить исходный столбец Address Попробуйте решить эту задачу самостоятельно, а затем можете заглянуть в решение

```python
import pandas as pd
customers = pd.read_csv("customers.csv")
customers.head()
```

```python
# expand - возвращаем DataFrame
addr = customers['Address'].str.split(',', expand = True)
addr.head()
```

```python
customers[['Street', 'City', 'State', 'Zip']] = addr
customers.head()
```

```python
customers = customers.drop('Address', axis=1)
customers.head()
```

```python
customers[['First Name', 'Last Name']] = customers['Name'].str.split(pat = " ", expand = True, n =1)
customers.head()
```

```python
customers = customers.drop('Name', axis=1)
customers.head()
```
