```python

data = {'user' : ['Ivan', 'Petr', 'Nikolay', 'Andrey'],
            'time_per_page' :  [10, 30 , 15, 5],
             'pages' : [7, 9, 12, 3]}
df = pd.DataFrame(data, index=['user1','user2','user3','user4'])
s1 = pd.Series([200,150], index=['user3','user4'])

def solution(df, s1):
    df['balance'] = pd.Series(s1, index=['user3','user4'])
    return df
df1 = solution(df,s1)
df1
```

```python
df1[df1['balance'].isnull()].index
```
