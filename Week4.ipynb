Сначала выполним задание пункта "Задание для самопроверки 3". Данные для него доступны по [ссылке](https://studio.openedu.ru/assets/courseware/v1/b6cb4ea6a069a5a588596dd9e12de293/asset-v1:ITMOUniversity+DATANTECH2035+cifru_2035+type@asset+block/auto-mpg-quiz.csv).


Для начала подключаем библиотеку pandas и загружаем интересующий нас файл при помощи метода `read_csv(file_name, delimeter, index_col, decimal, usecols, header, ...)`. Подробное описание можно найти по [ссылке](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_csv.html).

````python
import pandas as pd
all_cars = pd.read_csv('https://studio.openedu.ru/assets/courseware/v1/b6cb4ea6a069a5a588596dd9e12de293/asset-v1:ITMOUniversity+DATANTECH2035+cifru_2035+type@asset+block/auto-mpg-quiz.csv',
                       delimiter=',',
                       decimal='.',
                       index_col = 'name')
````

Посмотрим на наши данные. Похоже, они загрузились правильно.

````python
all_cars.head()
````

Удалим те колонки, которые нам не потребуются (согласно заданию).

````python
all_cars = all_cars.drop(['origin', 'mpg', 'displ', 'yr', 'origin'], axis = 1)
all_cars.head()
````

На самом деле эти же колонки можно было не загружать сразу, используя, например, такой код:
```
all_cars = pd.read_csv('auto-mpg-quiz.csv',
                       delimiter=',',
                       index_col='name',
                       decimal='.',
                       usecols=['name', 'hp', 'cyl', 'weight', 'accel'])
```





Проверьте самостоятельно, что в результаты выполнения такого фрагмента кода получится интересующий нас набор данных.

Теперь выполним первый пункт опроса. Нам нужно отобрать четырехцилиндровые автомобили (`cyl = 4`). Делается это, например, так:



````python
all_cars[all_cars.cyl == 4]
````

Обращаться к колонкам можно и следующим образом:

````python
all_cars[all_cars['cyl'] == 4]
````

Предположим теперь, что мы хотим найти автомобили с четырехцилиндровым двигателем и мощностью больше, чем $80$ (`cyl == 4 & hp > 80`). Тогда нам достаточно исполнить следующий код:

````python
all_cars[(all_cars.cyl == 4) & (all_cars.hp > 80)]
````

Теперь найдем те четырехцилиндровые машины, у которых мощность больше $80$, или те восьмицилиндровые, у которых мощность больше $90$. Логическое выражение может быть записано следующим образом:
```
((all_cars.cyl == 4) & (all_cars.hp > 80)) | ((all_cars.cyl == 8) & (all_cars.hp > 90))
```



Выполним соответствующий код и проверим результат:

````python
sample_cars = all_cars[((all_cars.cyl == 4) & (all_cars.hp > 80)) | ((all_cars.cyl == 8) & (all_cars.hp > 90))]
sample_cars
````

Определим среднее значение элементов столбца `weight`. Как обычно, возможны разные варианты обращения к колонкам набора данных:

````python
print(sample_cars['weight'].mean())
print(sample_cars.weight.mean())
````

Для округления можно пользоваться функцией `round(num, dec)`:



````python
round(sample_cars.weight.mean(), 3)
````
