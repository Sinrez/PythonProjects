```python
import pandas as pd
cars = pd.read_csv("used_cars.csv")
cars.head()
```

```python
min_wage = pd.read_csv("minimum_wage.csv")
min_wage.head()
```

```python
# Агрегируйте сумму цен машин в cars Сгруппируйте результаты по типу топлива по оси строк
cars.pivot_table(
    index = "Fuel", values = "Price", aggfunc = "sum"
)
```

```python
# Агрегируйте количество машин в cars
# Сгруппируйте результаты по произ- водителю на оси строк и типу коробки передач на оси столбцов
cars.pivot_table(
    index = "Manufacturer",
    columns = "Transmission",
    aggfunc = "count",
    margins = True,
    values = 'Price'
).head()
```

```python
# Агрегируйте среднюю цену машин в cars Сгруппируйте результаты по году
# и типу топлива на оси строк и типу коробки передач на оси столбцов
cars.pivot_table(
    index = ["Year",'Fuel'],
    columns = "Transmission",
    values = 'Price',
    aggfunc = "mean",
    fill_value = 0
)
```

```python
# В объекте DataFrame из предыдущей задачи перенесите уровень Transmission с оси столбцов на ось строк
cars_list = cars.pivot_table(
    index = ["Year",'Fuel'],
    columns = "Transmission",
    values = 'Price',
    aggfunc = "mean",
    fill_value = 0
)
cars_list.head()
```
