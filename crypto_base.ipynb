```python
from mod import Mod
# Funny math here

x = Mod(5, 7)      # x ≡ 5 (mod 7) то есть остаток от деления на 5 на 7 это 5

print((x + 2) == 0)      # True: 5 + 2 ≡ 7 ≡ 0 (mod 7) / 5 + 2 = 7 остаток от деления на 7 - это 0
print((x + 7) == x)       # True: 5 + 7 ≡ 12 ≡ 5 (mod 7)
print((x**3) == (x + 1))  # True: 5³ ≡ 125 ≡ 6 (mod 7)
print((1 // x) == 3 )    # True: 5 × 3 ≡ 15 ≡ 1 (mod 7) ⇒ 5⁻¹ ≡ 3 (mod 7)
```
