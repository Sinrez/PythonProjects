```python
# Объедините данные о продажах за две недели в один DataFrame Назначьте дан-
# ным из набора week1 ключ "Week 1", а данным из набора week2 — ключ "Week 2"

all_sales = pd.concat(objs= [week1, week2], keys = ["Week 1", "Week 2"])
all_sales
```

```python
# Найдите клиентов, которые посещали ресторан на каждой из рассматриваемых недель

week1.merge(right = week2, how = "inner", on = "Customer ID").head()
```

```python
week1.merge(
                right = week2, how = "inner", on = "Customer ID"
            ).drop_duplicates(subset = ["Customer ID"]).head()
```

```python
# Найдите клиентов, которые посещали ресторан
# на каждой рассматриваемой неделе и каждую неделю заказывали одно и то же блюдо
week1.merge(right = week2, how = "inner", on = ["Customer ID",'Food ID']).head()
```

```python
# Найдите клиентов, посещавших ресторан только на первой неделе и только на второй неделе
week1.merge(
                right = week2,
                how = "outer",
                on = "Customer ID",
                indicator = True
).head()
```

```python
# Каждая строка в наборе данных week1 идентифицирует клиента, заказавшего блюдо
# Для каждой строки в week1 извлеките информацию о клиенте из на- бора данных customers
week1.merge(right = customers,
            how = "left",
            left_on = "Customer ID",
            right_on = 'ID').head()
```
