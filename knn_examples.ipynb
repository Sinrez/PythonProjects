```python
import pandas as pd
import numpy as np
data = pd.DataFrame(np.array([
    [1, 28, 10,1],
    [2,	49,	49,	1],
    [3,	48,	35,	0],
    [4,	36,	33,	1],
    [5,	45,	54,	0]
]), columns=['id','X','Y','Class'])
data
```

```python
# Введите идентификатор (id) ближайшего соседа для объекта (33,47) , используя евклидову метрику.
from math import sqrt
x, y = 33,47
data['evkl'] = np.sqrt((data['X'] - x)**2 + (data['Y'] - y)**2)
#  манх
data['manh'] = abs(data.X-x)+abs(data.Y-y)
# Чебышев
data['che'] = abs(data.X-x)
data
```
