```python
import pandas as pd
tv_shows_json = pd.read_json("tv_shows.json")
tv_shows_json
```

```python
tv_shows_json.loc[0, "shows"]
```

Нормализуйте данные в каждом словаре в столбце "shows" В результате должен получиться DataFrame, в котором каждая серия хранится в отдельной записи Каждаязаписьдолжнавключатьсоответствующиеметаданныесерии (season, episode, name и air_date), а также информацию верхнего уровня о сериале (show, runtime и network)

```python
# {'show': 'The X-Files',
#  'runtime': 60,
#  'network': 'FOX',
#  'episodes': [{'season': 1,

tv_shows_df = pd.json_normalize(
    data = tv_shows_json['shows'],
    meta = ['show', 'runtime','network'],
    record_path = 'episodes'
)
tv_shows_df.head()
```

Разбейте нормализованный набор данных на три DataFrame, по одному для каждого сериала ("The X-Files", "Lost" и "Buffy the Vampire Slayer")

```python
xfiles = tv_shows_df[tv_shows_df["show"] == "The X-Files"]
lost = tv_shows_df[tv_shows_df["show"] == "Lost"]
buffy = tv_shows_df[tv_shows_df["show"] == "Buffy the Vampire Slayer"]
buffy.head()
```

Сохраните три DataFrame в книгу Excel episodes.xlsx, каждый в своем листе (имена листов выберите по своему вкусу)

```python
episodes = pd.ExcelWriter("episodes.xlsx")
episodes
```

```python
xfiles.to_excel(
                excel_writer = episodes, sheet_name = "X-Files", index = False
)
```

```python
lost.to_excel(
                excel_writer = episodes, sheet_name = "Lost", index = False
)
```

```python
 buffy.to_excel(
                excel_writer = episodes,
                sheet_name = "Buffy the Vampire Slayer",
                index = False
            )
```

```python
episodes.save()
```
